#### Dashboard to analyze the unemployment rate of each Indian state

```python
# plotting sunburst

import plotly.express as px

unemployment = df[['States','Region','Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region','States'],
                     values='Estimated Unemployment Rate',
                     width=700,height=600, color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```

```python
import plotly.express as px

unemployment = df[['States','Region','Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region','States','Estimated Unemployment Rate'],
                     values='Estimated Unemployment Rate',
                     width=1000,height=1000, color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```

```python
# plotting sunburst
import plotly.express as px

unemployment = df[['States', 'Region', 'Estimated Unemployment Rate']]

fig = px.treemap(unemployment, path=['Region', 'States'], values='Estimated Unemployment Rate',
                 width=800, height=800, color='Estimated Unemployment Rate',
                 color_continuous_scale='RdYlGn', title="Unemployment")

fig.show()
```
